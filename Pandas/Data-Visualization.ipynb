## 1. Bar Graph

```python
import pandas as pd
```

```python
tdf = pd.read_csv("Titanic_data.csv")
```

```python
tdf.head()
```

```python
tdf.Survived.value_counts().plot(
    kind = 'barh', 
    figsize=(4,3), 
    legend=True, 
    title='Passenger Survival Rate',
    colormap='cool'
)
```

```python
import matplotlib.pyplot as plt
```

```python
x = tdf.Survived.value_counts()
plt.bar(["Dead", "Survived"], x.to_numpy())
```

```python
x1 = tdf[['Sex', 'Survived']].value_counts()
plt.bar(["Dead", "Survived", "Dead", "Survived"], x1.to_numpy())
```

## 6. Heatmap

```python
tdf.dtypes
```

```python
s = tdf[['Pclass', 'Age', 'Fare', 'SibSp', 'Parch']].corr()
s
```

```python
import seaborn as sns
```

```python
sns.heatmap(s, cmap="crest", annot=True)
```

* Negative value represents Negative correlation
* Positive value represents Positive correleation
* Small value means minimal relation between the columns
