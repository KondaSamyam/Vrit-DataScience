### Inserting and Deleting columns and rows

```python
#Dropping Columns
new_df = tdf.drop('PassengerId', axis=1)
new_df
```

```python
new_df.head()
```

```python
#Deleting Rows
tdf.drop([0,2,4,6], axis=0, inplace=False)
```

```python
#Adding/Creating Columns
lastnames = [name.split(",")[0] for name in tdf.Name]
lastnames
```

```python
tdf['Lastname'] = lastnames
```

```python
tdf.head()
```

```python
s = [name.split(",")[1].split(".")[0].strip() for name in tdf.Name]
s
```

```python
tdf["Salutations"] = s
tdf
```

```python
tdf.drop(['Salutations', 'Lastname'], axis=1, inplace=True)
```
