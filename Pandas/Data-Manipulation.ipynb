### dataframe.apply()

```python
def get_salutations(fullname):
    firstname = fullname.split(",")[1]
    initials = firstname.split(".")[0]
    return initials.strip()
```

```python
tdf['initials'] = tdf.Name.apply(get_salutations)
tdf
```

```python
tdf.initials.value_counts()
```

```python
tdf.initials.value_counts().plot.bar()
```

```python
tdf.initials.replace(to_replace="Dr", value="Mr", inplace=True)
```

```python
tdf.initials.value_counts()
```

```python
tdf.Survived.value_counts()
```

```python
survived_dict = {
    0: "Died",
    1: "Alive"
}

rdf = tdf.Survived.replace(survived_dict, inplace=False)
```

```python
rdf.value_counts()
```

```python
init_dict = {
    "Ms":"Miss", "Mlle":"Miss", "Lady":"Miss","Sir":"Mr","Capt":"Mr","Rev":"Master","Major":"Mr","Mme":"Miss",
    "Don":"Mr", "Col":"Mr", "the Countess":"Mrs", "Jonkheer":"Mr"
}
```

```python
tdf.initials.replace(to_replace=init_dict, inplace=True)
```

```python
tdf.initials.value_counts()
```

```python
group_by_init = tdf.groupby(by="initials", axis=0)
```
